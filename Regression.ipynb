```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import AdaBoostRegressor, RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.svm import SVR
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_excel("/content/sel_maccs.xlsx")
X = df.loc[:, 'MACCS_0':'MACCS_166']
X.columns = X.columns.astype(str)
y = df ['logsi']
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=4
)
```

Linear Regression

```python
lin_regressor = LinearRegression()
model = lin_regressor.fit(X_train, y_train)
y_pred = model.predict(X_test)
print("Score: ", r2_score(y_test,y_pred))
print("RMSE: ",mean_squared_error(y_test, y_pred))
```

Decision Tree

```python
DT_regressor = DecisionTreeRegressor()
model = DT_regressor.fit(X_train, y_train)
y_pred = model.predict(X_test)
print("Score: ", r2_score(y_test,y_pred))
print("RMSE: ",mean_squared_error(y_test, y_pred))
```

Random Forrest

```python
RF_regressor = RandomForestRegressor()
model = RF_regressor.fit(X_train, y_train)
y_pred = model.predict(X_test)
print("Score: ", r2_score(y_test,y_pred))
print("RMSE: ",mean_squared_error(y_test, y_pred))
```

Support Vector Machine

```python
SV_regressor = SVR()
model = SV_regressor.fit(X_train, y_train)
y_pred = model.predict(X_test)
print("Score: ", r2_score(y_test,y_pred))
print("RMSE: ",mean_squared_error(y_test, y_pred))
```

Adaboost

```python
AD_regressor = AdaBoostRegressor()
model = AD_regressor.fit(X_train, y_train)
y_pred = model.predict(X_test)
print("Score: ", r2_score(y_test,y_pred))
print("RMSE: ",mean_squared_error(y_test, y_pred))
```

XGBoost

```python
XG_regressor = XGBRegressor()
model = XG_regressor.fit(X_train, y_train)
y_pred = model.predict(X_test)
print("Score: ", r2_score(y_test,y_pred))
print("RMSE: ",mean_squared_error(y_test, y_pred))
```
