```python
import umap
import pandas as pd
import plotly.express as px


df = pd.read_excel("/content/maccs.xlsx")
fp = df.loc[:, 0:166]

umap = umap.UMAP(n_components=3, init='random', random_state=0)

proj_3d = umap.fit_transform(fp)

fig = px.scatter_3d(
   proj_3d, x=0, y=1, z=2,
   color=df.Activity, labels={'color': 'Activity'}
)

fig.show()
fig.write_image('maccs.png')
```
